```python
#fixing "1 to 3" for the number/percentage columns
#change these vars and run code
edited_file = "/content/drive/MyDrive/DS3010/csv/2013-2014/2013_2014_Retention.csv"
num_col_name = "Number_English Language Learners"
percent_col_name = "Percent_English Language Learners"

df = pd.read_csv(edited_file)

df["Total Students"] = (
    df["Total Students"]
    .astype(str)
    .str.replace(",", "", regex=False)
    .str.strip()
    .astype(int)
)

mask = (
    (df[num_col_name] == "1 to 3") &
    (df[percent_col_name] != "1 to 3")
)

df.loc[mask, num_col_name] = (
    (
        df.loc[mask, percent_col_name].astype(float) / 100
    ) * df.loc[mask, "Total Students"]
).round().astype(int)

df.to_csv(edited_file, index=False)
```

```python
#make everything an integer/float that is not already
edited_file = "/content/drive/MyDrive/DS3010/csv/2015-2016/2015_2016_Retention.csv"

df = pd.read_csv(edited_file)

int_columns = [
    "Total Students",
    "Number_Male",
    "Number_Female",
    "Number_American Indian or Alaska Native",
    "Number_Asian",
    "Number_Hispanic or Latino of any race",
    "Number_Black or African American",
    "Number_White",
    "Number_Native Hawaiian or Other Pacific Islander",
    "Number_Two or more races",
    "Number_Students With Disabilities Served Under IDEA",
    "Number_Students With Disabilities Served Only Under Section 504",
    "Number_English Language Learners",
    "Number of Schools",
    "Start year"
  ]

float_columns = [
    "Percent_American Indian or\nAlaska Native",
    "Percent_Asian",
    "Percent_Hispanic or Latino of any race",
    "Percent_Black or African American",
    "Percent_White",
    "Percent_Native Hawaiian or Other Pacific Islander",
    "Percent_Two or more races",
    "Percent_Students With Disabilities Served Under IDEA",
    "Percent_Students With Disabilities Served Only Under Section 504",
    "Percent_English Language Learners",
    "Percent of Schools Reporting ",
  ]

for col in int_columns:
    df[col] = df[col].astype(str).str.replace(",", "", regex=False).str.strip()

    mask = df[col] != "1 to 3"

    df.loc[mask, col] = df.loc[mask, col].astype(int)

for col in float_columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

df.to_csv(edited_file, index=False)
```

```python
#replace any values in male and female that equal "1 to 3"
import numpy as np

edited_file = "/content/drive/MyDrive/DS3010/csv/2013-2014/2013_2014_Retention.csv"
df = pd.read_csv(edited_file)

male_col = "Number_Male"
female_col = "Number_Female"
total_col = "Total Students"

male_mask = df[male_col] == "1 to 3"
female_mask = df[female_col] == "1 to 3"
both_mask = male_mask & female_mask

# convert needed columns safely
df[total_col] = pd.to_numeric(df[total_col].astype(str).str.replace(",", ""), errors="coerce")
df[male_col] = pd.to_numeric(df[male_col], errors="coerce")
df[female_col] = pd.to_numeric(df[female_col], errors="coerce")

# skip rows where both are "1 to 3"
valid_male = male_mask & ~both_mask
valid_female = female_mask & ~both_mask

# replace male = total - female
df.loc[valid_male, male_col] = df.loc[valid_male, total_col] - df.loc[valid_male, female_col]

# replace female = total - male
df.loc[valid_female, female_col] = df.loc[valid_female, total_col] - df.loc[valid_female, male_col]

df.to_csv(edited_file, index=False)
```
